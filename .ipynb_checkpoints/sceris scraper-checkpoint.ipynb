```python
import os
import time
import datetime
from selenium import webdriver

url = 'https://scerisecm.boston.gov/ScerIS/CmPublic/#/Home' # base

# for browser interaction & load waiting:
from selenium.common.exceptions import TimeoutException
from selenium.webdriver.support.ui import WebDriverWait, Select
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
```

```python
# returns target element once it is loaded
def wait_for(element, by, timeout=10):
    try:
        WebDriverWait(d, timeout).until(
            EC.presence_of_element_located((by, element)))
    except TimeoutException: print(f"Timeout limit reached ({timeout} s)")
    finally: return d.find_element(by, element)
```

```python
# scrapes for short but fast proxies list
def get_proxies():
    driver = webdriver.Chrome()
    driver.get('https://hidemy.name/en/proxy-list/?maxtime=600&type=s&anon=34#list')
    table = driver.find_element_by_class_name('table_block')
    proxies = []
    for tr in table.find_elements_by_tag_name('tr')[1:]:
        proxy = []
        for td in tr.find_elements_by_tag_name('td')[:2]:
            proxy.append(td.text.strip())
        proxies.append(':'.join(proxy))
    driver.quit()
    return proxies
proxies = get_proxies()
print('> Acquired', len(proxies), 'proxies')
print('—'*30)
```

# create time range index

```python
base = datetime.datetime.strptime(deepest_so_far, '%m/%d/%Y')

# base = datetime.datetime.today()

numdays = 365*2 # last two years
date_list = [base - datetime.timedelta(days=x) for x in range(numdays)]
date_list = [d.strftime('%m/%d/%Y') for d in date_list]
```

# scrape

```python

# universal element identifiers
url = 'https://scerisecm.boston.gov/ScerIS/CmPublic/#/Home'
open_search = '/html/body/div[1]/div/div/div[2]/div[3]/search-dashboard/div/div/div[2]/div/div/div/div/ul/li[3]/dashboard-folders-renderer/div/div/div/a[3]'
date_input = "input[datepicker-options='dateOptions']"
search = 'btnSearch'
downloader = "button[ng-click='exportSearchResult()']"
download = 'btnDownload'
back = "button[ng-click='back()']"

# launch driver, enlarge window
d = webdriver.Chrome()
d.set_window_position(0, 0)
d.set_window_size(1280, 1000)

# load page
d.get(url)
for date in date_list:

    # open search panel
    wait_for(open_search, By.XPATH).click()
    
    # input date param
    wait_for(date_input, By.CSS_SELECTOR).send_keys(date)
    
    # load results
    wait_for(search, By.ID).click()
    
    # check for any records
    wait_for('ui-grid-viewport', By.CLASS_NAME)
    records = [t.text for t in
           d.find_elements_by_class_name('ui-grid-viewport')]
    if 'No records found.' in ''.join(records):
        print(f'{date} - no records found.')
    
    else: # (if records)
        # open download link
        wait_for(downloader, By.CSS_SELECTOR).click()

        # download file
        wait_for(download, By.ID).click()
        deepest_so_far = date
        print(f'{date} - scraped')

    # return to search
    wait_for(back, By.CSS_SELECTOR).click()
    
    # sleep
    if int(date[:2])%10==0: time.sleep(3)
    else: time.sleep(1)
```
